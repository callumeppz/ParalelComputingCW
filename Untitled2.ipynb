```python
from pyspark.sql import SparkSession
import random
import matplotlib.pyplot as plt

def estimate_pi_spark(num_darts, num_partitions):
    """
    Estimates Pi using a Monte Carlo method with Spark.

    Args:
    - num_darts (int): Number of darts to simulate.
    - num_partitions (int): Number of partitions for parallel computation.

    Returns:
    - float: Estimated value of Pi.
    """

    # Initializing a Spark session
    spark = SparkSession.builder.appName("EstimatePi").getOrCreate()

    # Function to determine if a point is inside the unit circle quadrant
    def is_inside_circle(point):
        x, y = point
        x2, y2 = x ** 2, y ** 2
        return 1 if x2 + y2 <= 1 else 0

    # Generating random points within the unit square
    points = [(random.random(), random.random()) for _ in range(num_darts)]

    # Creating an RDD from the list of points with specified partitions
    rdd_points = spark.sparkContext.parallelize(points, num_partitions)

    # Counting the points inside the unit circle quadrant using map-reduce
    inside_circle = rdd_points.map(is_inside_circle).reduce(lambda x, y: x + y)

    # Calculating Pi estimate
    pi_estimate = 4 * inside_circle / num_darts

    # Stopping the Spark session
    spark.stop()

    return pi_estimate

def plot_accuracy_vs_darts_spark(num_partitions):
    """
    Plots Pi accuracy against the number of darts for fixed partitions.

    Args:
    - num_partitions (int): Number of partitions for Spark computation.
    """

    num_darts_list = [100, 1000, 5000, 10000]
    pi_estimates = []

    for num_darts in num_darts_list:
        pi = estimate_pi_spark(num_darts, num_partitions)
        pi_estimates.append(pi)

    plt.plot(num_darts_list, pi_estimates, marker='o', label=f'Partitions={num_partitions}')
    plt.xlabel('Number of Darts')
    plt.ylabel('Estimated Pi')
    plt.title(f'Accuracy of Pi vs. Number of Darts (Partitions={num_partitions})')
    plt.legend()
    plt.show()

def plot_accuracy_vs_partitions_spark(num_darts):
    """
    Plots Pi accuracy against the number of partitions for fixed darts.

    Args:
    - num_darts (int): Number of darts for estimation.
    """

    num_partitions_list = [2, 4, 6, 8]
    pi_estimates = []

    for num_partitions in num_partitions_list:
        pi = estimate_pi_spark(num_darts, num_partitions)
        pi_estimates.append(pi)

    plt.plot(num_partitions_list, pi_estimates, marker='o', label=f'Darts={num_darts}')
    plt.xlabel('Number of Partitions (Nodes)')
    plt.ylabel('Estimated Pi')
    plt.title(f'Accuracy of Pi vs. Number of Partitions (Darts={num_darts})')
    plt.legend()
    plt.show()

def scaling_discussion():
    """
    Discusses scaling observations and unexpected behaviors.
    """

    print("Scaling Observations:")
    # Discuss scaling observations, trends, or unexpected behaviors observed.
    # Explain the impact of changing the number of darts and partitions on Pi accuracy.

# Example usage:
num_partitions_input = 4  # Change this value to observe different partition scenarios
num_darts_input = 10000  # Change this value for different dart throwing scenarios

plot_accuracy_vs_darts_spark(num_partitions_input)
plot_accuracy_vs_partitions_spark(num_darts_input)
scaling_discussion()
```
